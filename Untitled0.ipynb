```python
#handling  missing data values
#scenario1:where the numerical values is missing in the cloumn
#scenario2:where the word in a particualr row is being missed
#scenario3:wgere the data is in the huge quantity and the missing values are very less

import numpy as np
from sklearn.impute import SimpleImputer
#simpleImpurer is a class which was imported from sklearn  and is used to fill the missing values in the data
#OneHotEncoder is used for encoding of input values
#LinearEncoder is used to encode the output data values.
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit(x[:,1:3])
x[:,1:3]=imputer.transform(x[:,1:3])
```

```python
x
```

```python
#finding the catagical datat data into numerical data
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
#converting the cateogorical texlumnTransformer
#label encode is a class which was imported from sklearn.preprocessing. it is used to convert text data into binary data
#coloumn transformer is a clas importedfrom sklearn.compose. it is used to apply changes/ transformers to columns into array.it is used to apply changes to numercal data into categorical data.
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],remainder='passthrough')
#asthe data to be prossed , is in the 0th row the zero is mentioned
x=ct.fit_transform(x)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
#step5: split the data into training and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
#  it was usef for splitng the data into traianing and testing
```
